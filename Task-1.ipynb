# Exploring and visualizing Dataset


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  
```

```python
data = sns.load_dataset("iris")
```

```python
df = pd.DataFrame(data)
```

```python
df.describe() #this show data min max ,std and values
```

```python
df.isnull().sum()
```

```python
df_shape = df.shape # to know how many total rows and columns in our dataset
```

```python
data_boxplot = sns.boxplot(df) # this is the boxplot to define outliers of whole data
plt.savefig("data_boxplot.png")
```

```python
data_pairplot = sns.pairplot(df)
plt.savefig("data_pairplot.png")
```

```python
data_outliers = sns.boxplot(x="species", y="sepal_length", data=df) # this charts shows outliers in our data set
plt.savefig("data_outliers.png")
```

```python
data_hist = plt.hist(df["petal_length"], bins=20,histtype="step")
plt.savefig("data_hist.png")
```

```python
data_scatter = sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.savefig("data_scatter.png")
```
